```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import normalize
plt.rcParams.update({'font.size': 14})
```

```python
df = pd.read_csv('phobert-base-v2-epoch6_val.csv')
```

```python
cm = confusion_matrix(df['sentiment'], df['sentiment_pred'])

cm_normalized = normalize(cm, norm='l1', axis=1)

# Create ConfusionMatrixDisplay objects
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp_normalized = ConfusionMatrixDisplay(confusion_matrix=cm_normalized)

# Plot the confusion matrices
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

disp.plot(ax=ax1, cmap=plt.cm.Blues)
ax1.set_title('Confusion Matrix', fontsize=18)
ax1.set_xlabel('Predicted label', fontsize=16)
ax1.set_ylabel('True label', fontsize=16)

disp_normalized.plot(ax=ax2, cmap=plt.cm.Blues)
ax2.set_title('Normalized Confusion Matrix', fontsize=18)
ax2.set_xlabel('Predicted label', fontsize=16)
ax2.set_ylabel('True label', fontsize=16)

plt.tight_layout()
plt.show()
```

```python
cm = confusion_matrix(df['classification'], df['classification_pred'])
plt.rcParams.update({'font.size': 10})
cm_normalized = normalize(cm, norm='l1', axis=1)

# Create ConfusionMatrixDisplay objects
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp_normalized = ConfusionMatrixDisplay(confusion_matrix=cm_normalized)

# Plot the confusion matrices
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

disp.plot(ax=ax1, cmap=plt.cm.Blues)
ax1.set_title('Confusion Matrix', fontsize=18)
ax1.set_xlabel('Predicted label', fontsize=16)
ax1.set_ylabel('True label', fontsize=16)

disp_normalized.plot(ax=ax2, cmap=plt.cm.Blues,values_format='.2f')
ax2.set_title('Normalized Confusion Matrix', fontsize=18)
ax2.set_xlabel('Predicted label', fontsize=16)
ax2.set_ylabel('True label', fontsize=16)

plt.tight_layout()
plt.show()
```
